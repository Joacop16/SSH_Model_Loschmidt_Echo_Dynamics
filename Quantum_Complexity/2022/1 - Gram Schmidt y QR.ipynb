## 1. Gram-Schmidt

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import nbimporter
import Funciones
```

Intentemos construir la base de Krylov diagonalizandola directamente usando Gram-Schmit a ver que sucede:

```python
N = 10 #Intentemos con N pequeños solo para ver que sucede
t1 = 0.5
t2 = 1
D, U, H = Funciones.Diagonalizar(Funciones.H_t1_t2(N, t1,t2))
Vec_Est_Base, lambda_Base = Funciones.Estado_Base(D,U)
Krylov_Basis = Funciones.Krylov_Basis(H, Vec_Est_Base)
```

```python
pd.DataFrame(Krylov_Basis)
```

```python
def Gram_Schmidt(Krylov_Basis):
    N = Krylov_Basis.shape[0]
    for n in range(N):
        for i in range(n):
            Krylov_Basis[:,n] = Krylov_Basis[:,n] - float(np.transpose(Krylov_Basis[:,n])@Krylov_Basis[:,i])*Krylov_Basis[:,i]
            #print(f'Krylov_Basis[:,{n}] = Krylov_Basis[:,{n}] - float(np.transpose(Krylov_Basis[:,{n}])@Krylov_Basis[:,{i}])*Krylov_Basis[:,{i}]')
        if (np.isclose(np.linalg.norm(Krylov_Basis[:,n]),0)): print(f'el vector Krylov_Basis[:,{n}] es nulo')
        else: Krylov_Basis[:,n] = Krylov_Basis[:,n]/np.linalg.norm(Krylov_Basis[:,n])

    return Krylov_Basis
```

```python
Krylov_Basis = Gram_Schmidt(Krylov_Basis)
```

```python
pd.DataFrame(Krylov_Basis)
```

Practicamente es una matriz de ceros, esto sucede debido a que si el estado inicial es un vector propio entonces al aplicar H solo se generaran vectores linealmente dependientes, es imposible construir algo ortogonal apartir de esto.

```python
N = 100 #Intentemos con N pequeños solo para ver que sucede
t1 = 0.5
t2 = 1
D, U, H = Funciones.Diagonalizar(Funciones.H_t1_t2(N, t1,t2))

u_0 = np.transpose(np.array([np.random.random(N)]))
Krylov_Basis = Funciones.Krylov_Basis(H, u_0)
Krylov_Basis = Gram_Schmidt(Krylov_Basis)
```

```python
pd.DataFrame(Krylov_Basis)
```

```python
Funciones.revisar_ortogonalidad(Krylov_Basis)
```

```python
pd.DataFrame(np.linalg.inv(Krylov_Basis)@H@Krylov_Basis)
```

A simple vista pareciera que funciona. Extraígamos los coeficientes de Lanczos:

```python
def a_b_coefs(Krylov_Basis, H):
    tridiagonal = np.linalg.inv(Krylov_Basis)@H@Krylov_Basis
    a_coef = np.array([])
    b_coef = np.array([0])
    
    for i in range(H.shape[0]):
        a_coef = np.append(a_coef,tridiagonal[i,i])
        
    for i in range(H.shape[0]-1):
        if(np.isclose(tridiagonal[i,i+1],tridiagonal[i+1,i])): b_coef = np.append(b_coef,tridiagonal[i,i+1]) 
        else: print(f'Error, la matriz no es tridiagonal en las componentes [{i},{i+1}]')
        
    return a_coef, b_coef
```

```python
a_coef, b_coef = a_b_coefs(Krylov_Basis, H)
```

Aunque a simple vista parecia que era tridiagonal, desde la fila 34 dejo de serlo.

## 2. QR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy

import nbimporter
import Funciones
```

Intentemos construir la base de Krylov diagonalizandola directamente usando factorización QR a ver que sucede:

```python
N = 10 #Intentemos con N pequeños solo para ver que sucede
t1 = 0.5
t2 = 1
D, U, H = Funciones.Diagonalizar(Funciones.H_t1_t2(N, t1,t2))
Vec_Est_Base, lambda_Base = Funciones.Estado_Base(D,U)
Krylov_Basis = Funciones.Krylov_Basis(H, Vec_Est_Base)
```

```python
pd.DataFrame(Krylov_Basis)
```

```python
Q,R = scipy.linalg.qr(Krylov_Basis)
```

```python
Krylov_Basis = Q
```

```python
pd.DataFrame(Krylov_Basis)
```

Practicamente es una matriz de ceros, creo que esto sucede porque si construimos la base de Krylov usando el estado base, este sera un vector columna cada vez más cercano al cero.

```python
Funciones.revisar_ortogonalidad(Krylov_Basis)
```

```python
pd.DataFrame(np.linalg.inv(Krylov_Basis)@H@Krylov_Basis)
```

Si miramos bien, esta no es una matriz tridiagonal.

Sin embargo, en este caso como tal usamos un conjunto de vectores linealmente dependientes, probemos usando el estado inicial como un vector al azar:

```python
N = 100 #Intentemos con N pequeños solo para ver que sucede
t1 = 0.5
t2 = 1
D, U, H = Funciones.Diagonalizar(Funciones.H_t1_t2(N, t1,t2))

u_0 = np.transpose(np.array([np.random.random(N)]))
Krylov_Basis = Funciones.Krylov_Basis(H, u_0)
Q,R = scipy.linalg.qr(Krylov_Basis)
Krylov_Basis = Q
```

```python
pd.DataFrame(np.linalg.inv(Krylov_Basis)@H@Krylov_Basis)
```

Sucede lo mismo, aunque parece tridiagonal, hay un punto donde empieza a tener ceros fuera de la diagonal.
